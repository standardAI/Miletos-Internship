```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dims = [4096] * 15
hs = []
x = np.random.randn(16, dims[0])
for Din, Dout in zip(dims[:-1], dims[1:]):
    #W = np.random.randn(Din, Dout) / np.sqrt(Din)
    W = 0.022 * np.random.randn(Din, Dout)
    #x = 1/(1 + np.exp(-(x.dot(W))))
    x = np.tanh(x.dot(W))
    hs.append(x)
```

```python
#for h in hs:
#    print(h.mean(), h.std())
fig, axs = plt.subplots(1, 14, sharey=True, tight_layout=True, figsize=(20, 4))
for i, h in enumerate(hs):
    axs[i].hist(h.ravel(), bins=100, density=True)
    axs[i].set_title('std: %.4f' % h.std())
```
